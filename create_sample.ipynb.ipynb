```python
import pandas as pd

# Load full data
df = pd.read_csv('data/online_retail_II.csv', encoding='unicode_escape')

# Clean first
df = df.dropna(subset=['Customer ID'])
df = df[~df['Invoice'].str.startswith('C')]
df = df[df['Quantity'] > 0]
df = df[df['Price'] > 0]

# Get a RANDOM sample spread across all dates
sample = df.sample(n=10000, random_state=42)

# Save it
sample.to_csv('data/sample_data.csv', index=False)

print(f"Sample saved: {len(sample)} rows")
print(f"Date range: {sample['InvoiceDate'].min()} → {sample['InvoiceDate'].max()}")
print(f"Unique customers: {sample['Customer ID'].nunique()}")
```
